# Лабораторная работа №3 — Анализ результатов PINN (вариант 6)

ω₀ = 2.0, δ = 0.2, x(0) = 1.0, x'(0) = 0, T = 15. Слабое затухание.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

DATA = Path('..') / 'data'
FIG = Path('..') / 'figures'
```

```python
fwd = pd.read_csv(DATA / 'forward_solution.csv')
fwd.head()
```

```python
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(fwd['t'], fwd['x_analytical'], label='аналитическое')
ax.plot(fwd['t'], fwd['x_pinn'], '--', label='PINN')
ax.set_xlabel('t'); ax.set_ylabel('x(t)'); ax.grid(True, alpha=0.4); ax.legend()
plt.show()
```

```python
metrics_fwd = pd.read_csv(DATA / 'forward_metrics.csv')
metrics_inv = pd.read_csv(DATA / 'inverse_metrics.csv')
metrics_fwd, metrics_inv
```

```python
noisy = pd.read_csv(DATA / 'noisy_data.csv')
inv = pd.read_csv(DATA / 'inverse_solution.csv')
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(inv['t'], inv['x_true'], label='истинное')
ax.plot(inv['t'], inv['x_pinn'], '--', label='PINN')
ax.scatter(noisy['t'], noisy['y_noisy'], s=12, color='gray', alpha=0.7, label='зашумленные')
ax.set_xlabel('t'); ax.set_ylabel('x(t)'); ax.grid(True, alpha=0.4); ax.legend()
plt.show()
```

```python
hist_inv = pd.read_csv(DATA / 'inverse_history.csv')
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(hist_inv['epoch'], hist_inv['delta'])
ax.axhline(0.2, color='k', ls='--')
ax.set_xlabel('эпоха'); ax.set_ylabel('δ'); ax.grid(True, alpha=0.4)
plt.show()
```
