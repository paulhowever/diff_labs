# Lab2 Analysis (Variant 6)
Считывание RK4-результатов и визуализация ошибок.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

DATA_PATH = Path('..') / 'data' / 'rungekutta4.csv'
df = pd.read_csv(DATA_PATH)
df[['abs_error', 'rel_error']].max().to_frame('max_value')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['approx'], label='RK4 approx')
plt.plot(df['x'], df['exact'], 'k--', linewidth=2, label='Exact')
plt.title('Lab2: Exact vs RK4')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['x'], df['abs_error'], label='Absolute error')
plt.title('Lab2: Absolute Error')
plt.xlabel('x')
plt.ylabel('|exact - approx|')
plt.yscale('log')
plt.grid(True)
plt.legend()
plt.show()
```
